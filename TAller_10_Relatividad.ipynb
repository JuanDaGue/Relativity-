#1. 
Calcular la geodésica radial nula para la métrica de Schwarzschild, demuestre que los rayos de luz tiene una asintota en $r = 2m$.

#Solución
Sabemos que la métrica de Schwarchild es:
$$ds^2=\left(1-\frac{2m}{r}\right)dt^2-\left(1-\frac{2m}{r}\right)^{-1}ds^2-r^2\left(d\theta^2-\sin ^2 \varphi^2 \right )$$

Donde por simetría esférica tomamos $\dot \theta=\dot \varphi =0$, el cuerpo únicamente cambia su cordena radial y temporal.
$$ds^2=\left(1-\frac{2m}{r}\right)dt^2-\left(1-\frac{2m}{r}\right)^{-1}dr^2 )$$

**Rayo de Luz en la métrica de Schwarchild**

un rayo de luz recorre Distancias nulas con lo que tenemos:

$$0=\left(1-\frac{2m}{r}\right)dt^2-\left(1-\frac{2m}{r}\right)^{-1}dr^2$$
Reorganizando

\begin{eqnarray}
&\left(1-\frac{2m}{r}\right)dt^2&=&\left(1-\frac{2m}{r}\right)^{-1}dr^2\qquad(1.0)\\
\div \,\, d\tau ^2 &&&  \\
&\left(1-\frac{2m}{r}\right)\dot t^2&=&\left(1-\frac{2m}{r}\right)^{-1}\dot r^2\\
&\left(1-\frac{2m}{r}\right)^2\dot t^2&=&\dot r^2 \qquad(1.1)
\end{eqnarray}
donde $\dot t=\frac{dt}{d\tau}$ y $\dot r=\frac{dr}{d\tau}$

Si tomamos la ecuación $(1.0)$ e integramos para un rayo de luz que parte  desde una posición inicial $r_o$ a una posición arbitraria $r$ en un tiempo $t$ tenemos.

\begin{eqnarray}
dt&=&\pm \left(1-\frac{2m}{r}\right)^{-1}dr\\
\int_{t_0}^t dt&=&\pm \int_{r_0}^r  \left(1-\frac{2m}{r}\right)^{-1}dr\\
t&=&\pm \int_{r_0}^r\left(\frac{r}{r-2m}\right)dr\\
\end{eqnarray}
Haciendo un cambio de variable $u=r-2m$, $r=u+2m$ y $du=dr$ convertimos la integral en

\begin{eqnarray}
t&=&\pm \int_{u(r_0)}^{u(r)}\frac{u+2m}{u}du\\
t&=&\pm \left(u[_{u(r_0)}^{u(r)}+2m\ln(u)[_{u(r_0)}^{u(r)}\right)\\
t&=&\pm r-2m-r_o+2m+2m\ln\left(\frac{r-2m}{r_o-2m}\right)
\end{eqnarray}
tomamos la parte positiva 
$$t= r-r_o+2m\ln\left(\frac{r-2m}{r_o-2m}\right)$$
y tomamos la parte negativa 
$$t= r_o-r-2m\ln\left(\frac{r-2m}{r_o-2m}\right)$$
En esta ecuación se puede ver que es existe una asintota en $r=2m$, ya que logaritmo tiende a $- \infty$ , esto quiere decir que existe una singularidad en $r=2m$


```
import numpy as np
import matplotlib.pylab as plt

def tt (r,i,j):
  if r[i]< 2*m:
    t[i]=r0-r[i]-2*m*np.log((-2*m)/(r[i]-2*m))+t0[j]
  if r[i]>2*m:
    t[i]=r[i]-r0+2*m*np.log((r[i]-2*m)/(2*m))+t0[j]
  return t

m=10.0
r0=3.1*2*m*0.0
dt=10.0
#t0=np.array([dt*0.0,dt*1.0,dt*2,dt*3,dt*4,dt* 5.0,dt*6.0,dt*7.0,dt*8,dt*9,dt*10])
t0=np.arange(0.0,1000,10)
r=np.linspace(0,50,100  )
t=np.zeros(len(r))
Col=['b','r','black','b','r','g','black','r']

```

```
for i in range (len(r)):
  ts=tt(r,i,0)
plt.figure(figsize=(10,5))
for j in range(len(t0)):  
  plt.plot(r,ts+t0[j],Color=Col[2]) 
  plt.plot(-r+4*m,(-ts+t0[j]-50),Color=Col[3]) 
plt.plot(r*0+2*m,t,"darkred",label=r"Radio de Schwarchild $=2m$", linewidth=6)
plt.xlabel('r')
plt.ylabel('t')
plt.title('Rayo de luz en metrica de Swcharchild')
plt.ylim(0,50)
plt.xlim(0,40)
plt.grid()
plt.legend()
```

Se ve la métrica de Schwarchild para una masa de $10$ unidades

#Solución

derivando $\overline t =  t+2mln(r-2m)$ tenmos

$ {d\overline t} = dt+\frac{2m}{r-2m}dr$
despejando $dt$ y remmplazando en la metrica de scharchild 
$$ds^2=\left(1-\frac{2m}{r}\right)\left(\overline dt+\frac{2m}{r-2m}dr\right)^2-\left(1-\frac{2m}{r}\right)^{-1}dr^2- dl^2$$
donde $dl^2=r^2(d\theta+\sin ^2 \theta d\varphi)$  .
Expandiendo el termino cuadrático



$$ds^2=\left(1-\frac{2m}{r}\right)\left(d\overline t^2+\left(\frac{2m}{r-2m}\right)^2dr^2-\frac{4m}{r-2m}drd\overline t\right)-\left(1-\frac{2m}{r}\right)^{-1}dr^2- dl^2$$

Repartiendo los parentesis, y agrupando los elementos con $dr^2$

\begin{eqnarray}
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\left(\frac{r-2m}{r}\right)\frac{4m}{r-2m}drd\overline t-\left[\left(1-\frac{2m}{r}\right)\left(\frac{2m}{r-2m}\right)^2-\left(\frac{r-2m}{r}\right)^{-1}\right]dr^2- dl^2\\
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\frac{4m}{r}drd\overline t-\left[\frac{4m^2}{r(r-2m)}-\left(\frac{r-2m}{r}\right)^{-1}\right]dr^2- dl^2\\
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\frac{4m}{r}drd\overline t-\left[\frac{4m^2}{r(r-2m)}-\left(\frac{r}{r-2m}\right)\right]dr^2- dl^2\\
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\frac{4m}{r}drd\overline t-\left[\frac{4m^2-r^2}{r(r-2m)}\right]dr^2- dl^2\\
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\frac{4m}{r}drd\overline t-\left[\frac{(2m-r)(2m+r)}{r(r-2m)}\right]dr^2- dl^2\\
ds^2&=&\left(1-\frac{2m}{r}\right)d\overline t^2-\frac{4m}{r}drd\overline t-\left[\frac{(2m+r)}{r}\right]dr^2- dl^2\\
\end{eqnarray}


Con estas nuevas coordenadas llegamos a la métrica conocida como la métrica de $\texttt{Edington-Finkelstein}$

$$
(\mathrm{d} s)^{2}=\left(1-\frac{2m}{r}\right)\left(\mathrm{d} \overline t\right)^{2}- 2 \frac{2m}{r}  \mathrm{d} \overline t \mathrm{d} r-r^{2}\left(1+\frac{2m}{r}\right)(\mathrm{d} r)^{2}-\mathrm{d}\ell^2\qquad(5.0)
$$

para rayo luz en dirección que se mueve en dirección radial, la ecuación de las geodésicas nulas que caen en el agujero negro serán:

\begin{eqnarray}
\left(1-\frac{R_{\mathrm{S}}}{r}\right)\left(\mathrm{d} \overline t\right)^{2}- 2 \frac{2m}{r}  \mathrm{d} \overline t \mathrm{d} r-\left(1+\frac{2m}{r}\right)(\mathrm{d} r)^{2}=0
\end{eqnarray}

que se pueden escribir como:

\begin{eqnarray}
\left(1+\frac{2m}{r}\right)\left(\frac{\mathrm{d} r}{\mathrm{d} \overline t}\right)^{2}- 2 \frac{2m}{r} \frac{\mathrm{d} r}{\mathrm{d} \overline t}=\left(1-\frac{2m}{r}\right)
\end{eqnarray}

derivando $t^* =  t-2mln(r-2m)$ tenemos

$$ds^2=\left(1-\frac{2m}{r}\right)\left(dt^*+\frac{2m}{r-2m}dr\right)^2-\left(1-\frac{2m}{r}\right)^{-1}dr^2- dl^2$$
donde $dl^2=r^2(d\theta+\sin ^2 \theta d\varphi)$  .
Expandiendo el termino cuadrático



$$ds^2=\left(1-\frac{2m}{r}\right)\left(d t^{*2}+\left(\frac{2m}{r-2m}\right)^2dr^2+\frac{4m}{r-2m}drd t^*\right)-\left(1-\frac{2m}{r}\right)^{-1}dr^2- dl^2$$

Repartiendo los paren tisis, y agrupando los elementos con $dr^2$

\begin{eqnarray}
ds^2&=&\left(1-\frac{2m}{r}\right)d t^{*2}+\left(\frac{r-2m}{r}\right)\frac{4m}{r-2m}drd t^*-\left[\left(1-\frac{2m}{r}\right)\left(\frac{2m}{r-2m}\right)^2-\left(\frac{r-2m}{r}\right)^{-1}\right]dr^2- dl^2\\
